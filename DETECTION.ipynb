```python
li = ['lycon', 'cv2', 'PIL']
res = {k: {'read time': 0, 'save time':0, 'tot time': 0, 'resize':0} for k in li}
```

```python
import lycon
import time
r = time.time()
im1 = lycon.load('io.jpg')
res['lycon']['read time'] = time.time()-r
print(type(im1))
st = time.time()
lycon.save('io_saved.jpg', im1)
res['lycon']['save time'] = time.time()-st
rs = time.time()
lycon.resize(im1, width=299, height=299, interpolation=lycon.Interpolation.CUBIC)
res['lycon']['resize'] = time.time()-rs

```

```python
from PIL import Image
r = time.time()
im1 = Image.open('io.jpg')
res['PIL']['read time'] = time.time()-r
print(type(im1))
st = time.time()
im1.save('io_saved.jpg')
res['PIL']['save time'] = time.time()-st
print(time.time()-st)
```

```python
import cv2
r = time.time()
im1 = cv2.imread('io.jpg')
res['cv2']['read time'] = time.time()-r
print(type(im1))
st = time.time()
cv2.imwrite('io_saved.jpg', im1)
res['cv2']['save time'] = time.time()-st
print(time.time()-st)

rs = time.time()
cv2.resize(im1, dsize=(299, 299), interpolation=cv2.INTER_CUBIC)
res['cv2']['resize'] = time.time()-rs
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(res)
```
